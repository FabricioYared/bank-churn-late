## Script 2: Entrentamiento

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder
from sklearn.preprocessing import StandardScaler, MinMaxScaler, QuantileTransformer
from catboost import CatBoostClassifier
import pickle
```

```python
df = pd.read_csv("../data/processed/bank_train.csv")
```

```python
num_col = ['CreditScore', 'Age', 'Balance', 'EstimatedSalary']
obj_col = ['Geography', 'Gender']
```

```python
min_max = MinMaxScaler()
one_hot_encoder = OneHotEncoder()
```

```python
preprocessing = ColumnTransformer([
        ("num", min_max, num_col),
        ("obj", one_hot_encoder, obj_col),
        ('cat', 'passthrough', ['Tenure', 'NumOfProducts', 'HasCrCard', 'IsActiveMember'])
])
```

```python
X_train = df.drop(['Exited'], axis=1)
y_train = df[['Exited']]
```

```python
X_train1 = preprocessing.fit_transform(X_train)
```

```python
model = CatBoostClassifier()
```

```python
model.fit(X_train1, y_train)
```

```python
# Guardamos el modelo entrenado para usarlo en produccion
filename = '../models/best_model.pkl'
pickle.dump(model, open(filename, 'wb'))
```

```python
y_pred_test
```

```python

## Metricas de validación
def calc_metrics(y_test,y_pred_test):
    cm_test = confusion_matrix(y_test,y_pred_test)
    print("Matriz de confusion: ")
    print(cm_test)
    accuracy_test=accuracy_score(y_test,y_pred_test)
    print("Accuracy: ", accuracy_test)
    precision_test=precision_score(y_test,y_pred_test)
    print("Precision: ", precision_test)
    recall_test=recall_score(y_test,y_pred_test)
    print("Recall: ", recall_test)
```

```python
def save_plot(title):
    plt.title(title)
    fig = plt.gcf()
    filename = title.replace(" ", "_").lower()
    fig.savefig('{}'.format(filename), dpi=500)
    plt.clf()
```
